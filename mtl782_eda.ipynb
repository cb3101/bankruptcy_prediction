### Importing Libraries

```python
import os
from scipy.io import arff
import pandas as pd
import numpy as np
```

### Loading files

```python
path_to_directory= r"D:\mtl782_project\\"
files = [arff for arff in os.listdir(path_to_directory) if arff.endswith(".arff")]
print(files)
```

```python
df1 = arff.loadarff(r"D:\mtl782_project\\1year.arff")
data1 = pd.DataFrame(df1[0])
df2 = arff.loadarff(r"D:\mtl782_project\\2year.arff")
data2 = pd.DataFrame(df2[0])
df3 = arff.loadarff(r"D:\mtl782_project\\3year.arff")
data3 = pd.DataFrame(df3[0])
df4 = arff.loadarff(r"D:\mtl782_project\\4year.arff")
data4 = pd.DataFrame(df4[0])
df5 = arff.loadarff(r"D:\mtl782_project\\5year.arff")
data5 = pd.DataFrame(df5[0])
```

```python
for i in range(64):
    name = "Attr" + str(i+1)
    data1[name] = pd.to_numeric(data1[name], errors='coerce', downcast = "float")
    data2[name] = pd.to_numeric(data2[name], errors='coerce', downcast = "float")
    data3[name] = pd.to_numeric(data3[name], errors='coerce', downcast = "float")
    data4[name] = pd.to_numeric(data4[name], errors='coerce', downcast = "float")
    data5[name] = pd.to_numeric(data5[name], errors='coerce', downcast = "float")
```

## Replacing NA with Mean  

```python
for i in range(64):
    name = "Attr" + str(i+1)   
    data1[name] = data1[name].fillna(value= data1[name].mean())
    data2[name] = data2[name].fillna(value= data2[name].mean())
    data3[name] = data3[name].fillna(value= data3[name].mean())
    data4[name] = data4[name].fillna(value= data4[name].mean())
    data5[name] = data5[name].fillna(value= data5[name].mean())
```

```python
X1,y1 = data1.iloc[:,:-1], data1.iloc[:,-1]
X2,y2 = data2.iloc[:,:-1], data2.iloc[:,-1]
X3,y3 = data3.iloc[:,:-1], data3.iloc[:,-1]
X4,y4 = data4.iloc[:,:-1], data4.iloc[:,-1]
X5,y5 = data5.iloc[:,:-1], data5.iloc[:,-1]
```

```python
y1=y1.astype('int')
y2=y2.astype('int')
y3=y3.astype('int')
y4=y4.astype('int')
y5=y5.astype('int')
```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = ExtraTreesClassifier()

model.fit(X1,y1)
feat_importances = pd.Series(model.feature_importances_, index=X1.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("1 Year")
plt.show()

model.fit(X2,y2)
feat_importances = pd.Series(model.feature_importances_, index=X2.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("2 Year")
plt.show()

model.fit(X3,y3)
feat_importances = pd.Series(model.feature_importances_, index=X3.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("3 Year")
plt.show()

model.fit(X4,y4)
feat_importances = pd.Series(model.feature_importances_, index=X4.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("4 Year")
plt.show()

model.fit(X5,y5)
feat_importances = pd.Series(model.feature_importances_, index=X5.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("5 Year")
plt.show()



#print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
```
