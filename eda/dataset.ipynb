## Year 1

```python
arff = loadarff(data_path+'1year.arff')
df = pd.DataFrame(arff[0])
```

```python
df['class']= df['class'].astype('int')
df.head()
```

```python
df['class'].value_counts()
```

```python
df_pos = df[df['class']==1]
df_neg = df[df['class']==0]
```

```python
percent_missing = df_pos.isnull().sum()
percent_missing_p = percent_missing / df_pos.shape[0] * 100.00
missing_value_df = pd.DataFrame({'column_name': df_pos.columns,
                                 'num_missing': percent_missing,
                                 'percent_missing': percent_missing_p})
missing_value_df.sort_values('percent_missing', inplace=True, ascending=False)
missing_value_df[:10]
```

```python
percent_missing = df_neg.isnull().sum()
percent_missing_p = percent_missing / df_neg.shape[0] * 100.00
missing_value_df = pd.DataFrame({'column_name': df_neg.columns,
                                 'num_missing': percent_missing,
                                 'percent_missing': percent_missing_p})
missing_value_df.sort_values('percent_missing', inplace=True, ascending=False)
missing_value_df[:10]
```

```python
percent_missing = df.isnull().sum()
percent_missing_p = percent_missing / df.shape[0] * 100.00
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'num_missing': percent_missing,
                                 'percent_missing': percent_missing_p})
missing_value_df.sort_values('percent_missing', inplace=True, ascending=False)
missing_value_df[:10]
```

```python
with pd.option_context('display.max_columns', None):
    display(df_pos.describe())
    display(df_neg.describe())
    display(df.describe())
```
