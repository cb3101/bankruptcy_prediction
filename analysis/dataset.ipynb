```python
data_path = '../data/'
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io.arff import loadarff

%matplotlib inline
```

```python
arff1 = loadarff(data_path+'1year.arff')
df1 = pd.DataFrame(arff1[0])
```

```python
df1['class']= df1['class'].astype('int')
df1.head()
```

```python
df1['class'].value_counts()
```

```python
df1_pos = df1[df1['class']==1]
df1_neg = df1[df1['class']==0]
```

```python

percent_missing = df1_pos.isnull().sum()
percent_missing_p = percent_missing / df1_pos.shape[0] * 100.00
missing_value_df = pd.DataFrame({'column_name': df1_pos.columns,
                                 'num_missing': percent_missing,
                                 'percent_missing': percent_missing_p})
missing_value_df.sort_values('percent_missing', inplace=True, ascending=False)
missing_value_df
```

```python

percent_missing = df1_neg.isnull().sum()
percent_missing_p = percent_missing / df1_neg.shape[0] * 100.00
missing_value_df = pd.DataFrame({'column_name': df1_neg.columns,
                                 'num_missing': percent_missing,
                                 'percent_missing': percent_missing_p})
missing_value_df.sort_values('percent_missing', inplace=True, ascending=False)
missing_value_df
```
